```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import opendatasets as od
import os
```

```python
od.download("https://www.kaggle.com/datasets/sanikamal/rock-paper-scissors-dataset")
```

```python
import pandas as pd
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,      # Normalize pixel values
    rotation_range=20,   # Randomly rotate images
    width_shift_range=0.2, # Shift left/right
    height_shift_range=0.2,# Shift up/down
    shear_range=0.2,     # Warp the image
    zoom_range=0.2,      # Zoom in/out
    horizontal_flip=True,# Mirror the image
    fill_mode='nearest',
    validation_split=0.2 # Use 20% for testing
)
```

```python
data_dir = 'rock-paper-scissors-dataset/Rock-Paper-Scissors/train'
```

```python
train_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(150, 150), # Resize images to 150x150
    batch_size=32,
    class_mode='categorical',
    subset='training'
)
```

```python
validation_generator = train_datagen.flow_from_directory(
    data_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
```

```python
model = Sequential([
    # Layer 1: Detect Edges
    Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    MaxPooling2D(2, 2),

    # Layer 2: Detect Shapes (Fingers)
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # Layer 3: Detect Complex Objects (Hand poses)
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # Layer 4: Deep features
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),

    # Flatten: Turn the image into a list of numbers
    Flatten(),

    # The Brain: Making decisions
    Dropout(0.5), # Drop 50% of neurons to prevent overfitting
    Dense(512, activation='relu'),
    Dense(3, activation='softmax') # 3 Output Neurons: Rock, Paper, Scissors
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(
    train_generator,
    epochs=25, # Train for 25 rounds
    validation_data=validation_generator,
    verbose=1
)

# SAVE THE MODEL (Important!)
model.save("rps_model.h5")
```
