# Content Based recommending

```python
movies['metadata'] = movies['genres'].str.replace('|', ' ')
```

```python
print(movies[['title', 'metadata']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer(stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(movies['metadata'])
```

```python
tfidf_matrix[0]
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim
```

```python
def get_rec(movie_title, cosine_sim=cosine_sim):
  idx = movies.index[movies['title'] == movie_title][0]

  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:11]

  movie_indice = [i[0] for i in sim_scores]
  return movies[['title', 'metadata']].iloc[movie_indice]
```

```python
print(get_rec('Toy Story (1995)'))
```

# Content Based + titles

```python
# Create a 'soup' of words
def create_soup(x):
    # Combine Title and Genres into one string
    # We lower() everything so 'Action' and 'action' match
    return x['title'].lower() + ' ' + x['genres'].lower().replace('|', ' ')

movies['soup'] = movies.apply(create_soup, axis=1)

print(movies[['title', 'soup']].head())
```

```python
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['soup'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def get_hyb_rec(movie_title):
  idx = movies.index[movies['title'] == movie_title][0]

  sim_scores = list(enumerate(cosine_sim[idx]))

  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

  sim_scores = sim_scores[1:11]

  # 5. Return results
  movie_indices = [i[0] for i in sim_scores]
  return movies[['title', 'genres']].iloc[movie_indices]
```

# SVD recommendations

```python
import pandas as pd
import numpy as np
from scipy.sparse.linalg import svds
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
movies
```

```python
R_df = ratings.pivot(index='userId', columns='movieId', values='rating').fillna(0)
R_df
```

```python
R = R_df.values
R
```

```python
user_rating_mean = np.mean(R, axis=1)
user_rating_mean
```

```python
R_norm = R - user_rating_mean.reshape(-1, 1)
R_norm
```

```python
R.shape
```

```python
U, sigma, Vt = svds(R_norm, k=50)
print(U)
print(sigma)
print(Vt)
```

```python
sigma = np.diag(sigma)
sigma
```

```python
# 1. Multiply U * Sigma * Vt
# 2. Add the user averages back (Re-mean)
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_rating_mean.reshape(-1, 1)

# Turn into DataFrame for easy reading
preds_df = pd.DataFrame(all_user_predicted_ratings, columns=R_df.columns, index=R_df.index)

print("Predictions Generated for all 600+ users and 9000+ movies!")
```

```python
# Get the actual ratings (mask out the zeros)
actual_values = R[R.nonzero()]
predicted_values = all_user_predicted_ratings[R.nonzero()]



# Calculate RMSE
rmse = sqrt(mean_squared_error(actual_values, predicted_values))

print(f"Model RMSE: {rmse:.4f}")
# Logic: If RMSE is 0.8, it means our prediction is usually within 0.8 stars of the truth.
```

```python
def recommend_movies(user_id, preds_df, movies_df, original_ratings_df, num_recommendations=10):
    
    # 1. Get User's Predictions
    user_row_number = user_id - 1 
    sorted_user_predictions = preds_df.iloc[user_row_number].sort_values(ascending=False)
    
    # 2. Get User's History
    user_data = original_ratings_df[original_ratings_df.userId == user_id]
    already_seen = user_data['movieId'].tolist()
    
    # 3. Filter movies
    # Step A: Remove movies already seen
    recommendations = movies_df[~movies_df['movieId'].isin(already_seen)].copy()
    
    # Step B: Remove movies that have NO ratings (This fixes the KeyError!)
    # We only keep movies that exist as columns in our prediction matrix
    recommendations = recommendations[recommendations['movieId'].isin(preds_df.columns)]
    
    # 4. Add Predictions & Sort
    recommendations['Predicted_Rating'] = recommendations['movieId'].apply(lambda x: sorted_user_predictions[x])
    recommendations = recommendations.sort_values('Predicted_Rating', ascending=False)
    
    return recommendations.head(num_recommendations)

# --- TEST IT AGAIN ---
user_id = 432
recs = recommend_movies(user_id, preds_df, movies, ratings)

print(f"\n--- Top 10 SVD Recommendations for User {user_id} ---")
print(recs[['title', 'genres', 'Predicted_Rating']])
```

# SVD using Gradient Descent

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# ==========================================
# 1. LOAD & MAP DATA
# ==========================================

# We need to map Movie IDs and User IDs to a clean sequence (0, 1, 2, 3...)
# because the original IDs skip numbers (e.g., we have movie 1 and movie 5000)
user_ids = ratings['userId'].unique().tolist()
user_to_index = {x: i for i, x in enumerate(user_ids)}
index_to_user = {i: x for i, x in enumerate(user_ids)}

movie_ids = ratings['movieId'].unique().tolist()
movie_to_index = {x: i for i, x in enumerate(movie_ids)}
index_to_movie = {i: x for i, x in enumerate(movie_ids)}

ratings['user_idx'] = ratings['userId'].map(user_to_index)
ratings['movie_idx'] = ratings['movieId'].map(movie_to_index)

# Split Train/Test to measure accuracy
train, test = train_test_split(ratings, test_size=0.2, random_state=42)

# ==========================================
# 2. THE SGD ALGORITHM (The Netflix Solver)
# ==========================================
class SGD_MF:
    def __init__(self, n_users, n_items, n_factors=20, lr=0.01, reg=0.02):
        self.n_users = n_users
        self.n_items = n_items
        self.n_factors = n_factors
        self.lr = lr   # Learning Rate (How fast we learn)
        self.reg = reg # Regularization (Prevent over-memorizing)
        
        # Initialize Latent Factors with random noise
        # (Users Matrix and Movies Matrix)
        self.P = np.random.normal(scale=1./n_factors, size=(n_users, n_factors))
        self.Q = np.random.normal(scale=1./n_factors, size=(n_items, n_factors))
        
        # Initialize Biases (Some users just rate higher, some movies are just better)
        self.b_u = np.zeros(n_users)
        self.b_i = np.zeros(n_items)
        self.b = 0 # Global Average

    def fit(self, X, n_epochs=10):
        # Calculate global average
        self.b = np.mean(X[:, 2])
        
        print(f"Training on {len(X)} ratings for {n_epochs} epochs...")
        
        for epoch in range(n_epochs):
            # Shuffle data for randomness
            np.random.shuffle(X)
            
            sq_error = 0
            
            # THE LEARNING LOOP
            for u, i, r in X:
                u, i = int(u), int(i)
                
                # 1. Predict
                # Prediction = Global + UserBias + ItemBias + (UserFactors * ItemFactors)
                pred = self.b + self.b_u[u] + self.b_i[i] + np.dot(self.P[u], self.Q[i])
                
                # 2. Calculate Error
                err = r - pred
                sq_error += err**2
                
                # 3. Update Biases (Gradient Descent)
                self.b_u[u] += self.lr * (err - self.reg * self.b_u[u])
                self.b_i[i] += self.lr * (err - self.reg * self.b_i[i])
                
                # 4. Update Latent Factors
                # We want to move P and Q in the direction that reduces error
                p_u_old = self.P[u].copy() # Keep old value for Q update
                self.P[u] += self.lr * (err * self.Q[i] - self.reg * self.P[u])
                self.Q[i] += self.lr * (err * p_u_old - self.reg * self.Q[i])
            
            # Print RMSE every epoch
            rmse = np.sqrt(sq_error / len(X))
            print(f"Epoch {epoch+1}/{n_epochs} - RMSE: {rmse:.4f}")
            
    def predict(self, u_idx, i_idx):
        # Make a prediction for a single user/movie pair
        if u_idx >= self.n_users or i_idx >= self.n_items:
            return self.b # Default to average if unknown
        return self.b + self.b_u[u_idx] + self.b_i[i_idx] + np.dot(self.P[u_idx], self.Q[i_idx])

# ==========================================
# 3. RUN IT
# ==========================================
# Convert dataframe to numpy array [User, Movie, Rating]
train_data = train[['user_idx', 'movie_idx', 'rating']].values
test_data = test[['user_idx', 'movie_idx', 'rating']].values

# Initialize Model
model = SGD_MF(n_users=len(user_ids), n_items=len(movie_ids), n_factors=20, lr=0.01, reg=0.02)

# Train (Watch the RMSE drop!)
model.fit(train_data, n_epochs=10)

# ==========================================
# 4. EVALUATE ON HIDDEN DATA
# ==========================================
print("\nEvaluating on Test Set...")
errors = []
for u, i, r in test_data:
    pred = model.predict(int(u), int(i))
    errors.append((r - pred)**2)

final_rmse = np.sqrt(np.mean(errors))
print(f"Final Test RMSE: {final_rmse:.4f}")
```

```python
# Function to get recommendations for a user
def get_sgd_recommendations(user_id, model, movies_df, ratings_df, n=10):
    # Map User ID to Index
    u_idx = user_to_index[user_id]
    
    # Get all movie indices
    all_movie_indices = np.arange(len(movie_ids))
    
    # Identify movies already seen
    seen_movie_ids = ratings_df[ratings_df['userId'] == user_id]['movieId'].values
    seen_indices = [movie_to_index[mid] for mid in seen_movie_ids if mid in movie_to_index]
    
    # Filter out seen movies
    candidates = [i for i in all_movie_indices if i not in seen_indices]
    
    # Predict score for every candidate
    predictions = []
    for i_idx in candidates:
        pred_rating = model.predict(u_idx, i_idx)
        predictions.append((i_idx, pred_rating))
        
    # Sort by highest rating
    predictions.sort(key=lambda x: x[1], reverse=True)
    top_n = predictions[:n]
    
    # Convert back to Titles
    results = []
    for i_idx, score in top_n:
        original_movie_id = index_to_movie[i_idx]
        title = movies_df[movies_df['movieId'] == original_movie_id]['title'].values[0]
        results.append((title, score))
        
    return pd.DataFrame(results, columns=['Title', 'Predicted_Rating'])


# Recommend for User 432
print(get_sgd_recommendations(432, model, movies, ratings))
```
