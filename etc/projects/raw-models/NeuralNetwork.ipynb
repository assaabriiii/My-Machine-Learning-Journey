```python
import numpy as np
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivation(x):
  return x * (1 - x)
```

```python
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
```

```python
np.random.seed(42)
```

```python
# layer 1: inputs
```

```python
input_layer_neurons = 2
hidden_layer_neurons = 4
output_layer_neurons = 1

# Random weights (Heavier initialization helps)
weights_input_hidden = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
bias_hidden = np.random.uniform(size=(1, hidden_layer_neurons))

weights_hidden_output = np.random.uniform(size=(hidden_layer_neurons, output_layer_neurons))
bias_output = np.random.uniform(size=(1, output_layer_neurons))
```

```python
weights_input_hidden
```

```python
bias_hidden
```

```python
lr = 0.5
```

```python
print("training...")
for i in range(1000):

  # layer 1: inputs
  HLI = np.dot(X, weights_input_hidden) + bias_hidden
  HLO = sigmoid(HLI)

  # layer 2: hidden
  OLI = np.dot(HLO, weights_hidden_output) + bias_output
  PO = sigmoid(OLI)

  error = y - PO

  DPO = error * sigmoid_derivation(PO)

  EHL = DPO.dot(weights_hidden_output.T)

  DHL = EHL * sigmoid_derivation(HLO)

  weights_hidden_output += HLO.T.dot(DPO) * lr
  bias_output += np.sum(DPO, axis=0, keepdims=True) * lr

  # Update Layer 1 (FIXED: Used X instead of HLO)
  weights_input_hidden += X.T.dot(DHL) * lr
  bias_hidden += np.sum(DHL, axis=0, keepdims=True) * lr

  if i % 100:
    print(f"Epoch {i} Error: {np.mean(np.abs(error))}")

# 5. The Test
print("\nFinal Predictions:")
print(PO)
print("\nRounded:")
print(np.round(PO))
```
